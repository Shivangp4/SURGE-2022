### Interpolating r data and applying frozen in approximation

```python
from scipy import interpolate
index=np.linspace(0,2883,2884).astype(int)
x1=interpolate.interp1d(index,r_1[0],kind='linear',fill_value='extrapolate')
y1=interpolate.interp1d(index,r_1[1],kind='linear',fill_value='extrapolate')
z1=interpolate.interp1d(index,r_1[2],kind='linear',fill_value='extrapolate')
x2=interpolate.interp1d(index,r_2[0],kind='linear',fill_value='extrapolate')
y2=interpolate.interp1d(index,r_2[1],kind='linear',fill_value='extrapolate')
z2=interpolate.interp1d(index,r_2[2],kind='linear',fill_value='extrapolate')
x3=interpolate.interp1d(index,r_3[0],kind='linear',fill_value='extrapolate')
y3=interpolate.interp1d(index,r_3[1],kind='linear',fill_value='extrapolate')
z3=interpolate.interp1d(index,r_3[2],kind='linear',fill_value='extrapolate')
x4=interpolate.interp1d(index,r_4[0],kind='linear',fill_value='extrapolate')
y4=interpolate.interp1d(index,r_4[1],kind='linear',fill_value='extrapolate')
z4=interpolate.interp1d(index,r_4[2],kind='linear',fill_value='extrapolate')
```

```python
index=np.linspace(0,t1.shape[0]-1,t1.shape[0]).astype(int)
x_1=np.array(x1(index))-v1[0]*t1[0]
y_1=np.array(y1(index))-v1[1]*t1[0]
z_1=np.array(z1(index))-v1[2]*t1[0]
index=np.linspace(0,t1.shape[0]-1,t2.shape[0]).astype(int)
x_2=np.array(x2(index))-v2[0]*t2[0]
y_2=np.array(y2(index))-v2[1]*t2[0]
z_2=np.array(z2(index))-v2[2]*t2[0]
index=np.linspace(0,t1.shape[0]-1,t3.shape[0]).astype(int)
x_3=np.array(x3(index))-v3[0]*t3[0]
y_3=np.array(y3(index))-v3[1]*t3[0]
z_3=np.array(z3(index))-v3[2]*t3[0]
index=np.linspace(0,t1.shape[0]-1,t4.shape[0]).astype(int)
x_4=np.array(x4(index))-v4[0]*t4[0]
y_4=np.array(y4(index))-v4[1]*t4[0]
z_4=np.array(z4(index))-v4[2]*t4[0]
```

```python
X=np.concatenate((x_1,x_2,x_3,x_4))
Y=np.concatenate((y_1,y_2,y_3,y_4))
Z=np.concatenate((z_1,z_2,z_3,z_4))
```

```python
B=np.concatenate((B_1,B_2,B_3,B_4),axis=0)
B=B.transpose()
```

```python
bx=B[0]-B[0].mean()
by=B[1]-B[1].mean()
bz=B[2]-B[2].mean()
b=np.ndarray([3,bx.size])
b[0]=bx
b[1]=by
b[2]=bz
print(b)
r=np.sqrt(X**2+Y**2+Z**2)
```

```python
print(b)
plt.plot(r,b.T)
```

### Finding the overall discrete dataset and finding the correlation function

```python
index=np.linspace(np.min(r),np.max(r),1300000)
b_=interpolate.interp1d(r,b,kind='linear')
b_=np.array(b_(index))
```

```python
print(b_)
plt.plot(index,b_.T)
```

```python
r_max=999*(index[1]-index[0]) 
Rc=np.ndarray(1000) #correlation function
for i in range(1000):
    Rc[i]=np.mean(b_.T[:1300000-i]*b_.T[i:])
Rc=Rc/np.mean(b_*b_)
```

```python
# print(Rc)
plt.scatter(np.linspace(0,999,1000),Rc,s=0.4)
```

### Calculating the Taylor microscale from the correlation function and the technique described in Chuychai et al

```python

from scipy.optimize import curve_fit as cf
def f(x1,taylor):
    return (1-x1**2/(2*taylor**2))
  
    
ts2=np.ndarray(160)
j=0
for rmax in range(200,999,5):
    ts1=np.ndarray(rmax-99)
    rfit=np.linspace(100,rmax,rmax-99).astype(int)
    i=0
    for rfit1 in rfit:
        range_=np.linspace(0,rfit1,rfit1+1).astype(int)
        p_opt, p_cov = cf(f,range_,Rc[range_])
        ts1[i]=p_opt
        i=i+1
    plt.plot(rfit,ts1)
    plt.show()
    ts1_=interpolate.interp1d(rfit,ts1,kind='linear',fill_value='extrapolate')
    print(ts1_(0))
    ts2[j]=ts1_(0)
    j=j+1
print(ts2)
plt.plot(np.linspace(200,999,160),ts2)
ts2_=interpolate.interp1d(np.linspace(200,999,160),ts2,kind='linear',fill_value='extrapolate')
print(ts2_(0))
```
